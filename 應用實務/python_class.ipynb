```python
import PyPDF2
import urllib.request

pdfFiles = ["AboutReplIt.pdf", "UnknownPassword.pdf"]
for pdfFile in pdfFiles:
    urllib.request.urlretrieve("http://www1.pu.edu.tw/~yrjean/" + pdfFile, pdfFile)

# !pip install PyPDF2


def encryptPDF(fn):  # 加密fn PDF檔案
    pdfObj = open(fn, "rb")
    pdfRd = PyPDF2.PdfReader(pdfObj)
    pdfWr = PyPDF2.PdfWriter()  # Pdfwriter object is like a blank PDF file.
    # We need to add some pages to it before saving it to a file.

    for p in range(len(pdfRd.pages)):
        pdfWr.add_page(pdfRd.pages[p])

    pdfWr.encrypt("aabbcc")  # password: aabbec
    encryptPdf = open("Encrypted" + fn, "wb")
    pdfWr.write(encryptPdf)
    print("Encrypt " + fn + " and save encrypted output to Encrypted" + fn + "!")
    encryptPdf.close()


def readPDF(fn):  # 讀取fn PDF檔案，若是加密檔案，則破解之
    print("========== " + fn + " ==========")
    pdfObj = open(fn, "rb")
    pdfRd = PyPDF2.PdfReader(pdfObj)
    if pdfRd.is_encrypted:
        print("File " + fn + " is encrypted.")
        while True:
            password = input("Password:")
            if pdfRd.decrypt(password):
                print("+++ Decryption success!")
                pageObj = pdfRd.pages[0]
                txt = pageObj.extract_text()
                print("\nPage 1 : \n", txt + "\n")
                break
            else:
                print("+++ Decryption fail!")
                isStillContinue = input("+++ Try again?")
                if isStillContinue == "y":
                    continue
                else:
                    break
    else:
        pageObj = pdfRd.pages[0]
        txt = pageObj.extract_text()
        print("\nPage 1 : \n", txt + "\n")


print("=========================================")
encryptPDF("AboutReplIt.pdf")  # Encrypt it and generate
readPDF("AboutReplIt.pdf")  # No Encryption
readPDF("EncryptedAboutReplIt.pdf")  # Encrypted, password:aabbcc
readPDF("UnknownPassword.pdf")  # Encrypted, password unknown
```

### [HW0-2] 破解 PDF 密碼(3~4 位，數字、大小寫英文字母)


```python
import urllib.request
import PyPDF2

# !pip install PyPDF2
PDFfileName = input("PDF filename:")
urllib.request.urlretrieve(
    "http://www1.pu.edu.tw/~yrjean/" + PDFfileName + ".pdf", PDFfileName + ".pdf"
)


def fillWord():
    listword = []
    for i in range(0, 10):
        listword.append(str(i))
    for i in range(65, 91):
        listword.append(chr(i))
    for i in range(97, 123):
        listword.append(chr(i))
    return listword


def readPDF(fn):  # 讀取fn PDF檔案，若是加密檔案，則破解之
    words = fillWord()
    Decryption = False
    pdfObj = open(fn, "rb")
    pdfRd = PyPDF2.PdfReader(pdfObj)
    for p1 in words:
        for p2 in words:
            for p3 in words:
                Password = str(p1 + p2 + p3)
                if pdfRd.decrypt(Password):
                    Decryption = True
                    break
                for p4 in words:
                    Password = str(p1 + p2 + p3 + p4)
                    if pdfRd.decrypt(Password):
                        Decryption = True
                        break
                if Decryption == True:
                    break
            if Decryption == True:
                break
        if Decryption == True:
            break
    if Decryption == True:
        print("+++ Decryption success!")
        print("+++ Password is  " + Password)
    else:
        print("+++ Decryption fail!")


readPDF(PDFfileName + ".pdf")
```

### [EX1-3]Regular Expression


```python
import re

# A RE (regular expression) specifies a set of strings that matches it; the functions in this module let you check if a particular string matches a given regular expression
# 非從頭比對, 不match
print("----- match -----")  # '[a-z]+' : sequences of lower case letters
m = re.match("[a-z]+", "01tem56po")
print(m)

# 非從頭找起, 直到第一組吻合的sbstring
print("----- search -----")
m = re.search("[a-z]+", "01tem56po")
print(m)
print(m.group())  # returns the matched substring
print(m.span())
print(m.start())
print(m.end())

# 非從頭找起, 直到所有吻合的substring
print("----- findall -----")
m = re.findall("[a-z]+", "01tem56pa")
print(m)

print("----- findall -----")
m = re.findall("[a-zA-Z]+", "This is a book.")
print(m)

print("----- findall -----")
m = re.findall("[a-z'A-Z]+", "This isn't a book.")
print(m)

print("----- findall -----")
m = re.findall("[a-z'A-Z]+", "This isn't a book.")
print(m)
```

### [EX1-4]Regular Expression(Regular Expression Object)


```python
import re

# Regular Expression Object
pat1 = re.compile("[0-9]+")  # sequences of digits
pat2 = re.compile("0|[1-9][0-9]*")  # 零和非零開頭的數字，非負整數

# 從頭比對, 不match
print("----- match -----")
m = pat1.match("ABC0123DEF4")
print(m)

# 從頭找起, 直到第一組吻合的substring
print("----- search -----")
m = pat1.search("ABC0123DEF4")
print(m)
print(m.group())
print(m.span())

# 從頭找起, 直到所有吻合的substrings
print("----- pat1 findall -----")
m = pat1.findall("ABC0123DEF4")
print(m)

# 從頭找起, 直到所有吻合的substrings
print("----- pat2 findall -----")
m = pat2.findall("ABC0123DEF00407")
print(m)
```

### [HW2] SqliteManager: 帳密管理系統


```python
import sqlite3
import urllib.request
import os


def menu():
    print("=========================")
    print("帳密管理系統")
    print("-------------------------")
    print("1. 新增帳號")
    print("2. 顯示帳號")
    print("3. 修改密碼")
    print("4. 刪除帳號")
    print("0. 結束程式")
    print("-------------------------")


def choices(choice):
    if choice == "1":
        addNewAccount()
    elif choice == "2":
        showAllAccount()
    elif choice == "3":
        editAccount()
    elif choice == "4":
        deleteAccount()


def showAllAccount():
    cursor = conn.execute("select * from password")
    rows = cursor.fetchall()
    print("帳號  密碼")
    print("--------------")
    for row in rows:
        print("{}    {}".format(row[0], row[1]))
    print("--------------")
    input("按ENTER返回主選單")
    print()


def findSignalAccount(account, status):
    cursor = conn.execute('select * from password where name ="' + account + '"')
    data = cursor.fetchone()
    if status == "edit":
        print("原來密碼為：" + data[1])
    elif status == "checkAccountExist":
        if data is None:
            return False
        else:
            return True


def addNewAccount():
    pressEnterLeave = False
    while True:
        account = input("請輸入要新增的帳號(Enter==>停止輸入)")
        if account == "":
            pressEnterLeave = True
            break
        elif findSignalAccount(account, "checkAccountExist"):
            print(account + " 帳號已存在!")
        else:
            break
    if not pressEnterLeave:
        password = input("請輸入密碼：")
        add = 'insert into password values("' + account + '","' + password + '")'
        conn.execute(add)
        conn.commit()
        input("帳號新增完畢，請按ENTER返回主選單")
        print()


def editAccount():
    pressEnterLeave = False
    while True:
        account = input("請輸入要修改的帳號(Enter==>停止輸入)")
        if account == "":
            pressEnterLeave = True
            break
        elif findSignalAccount(account, "checkAccountExist"):
            break
        else:
            print(account + " 帳號不存在!")
    if not pressEnterLeave:
        findSignalAccount(account, "edit")
        password = input("請輸入密碼：")
        edit = 'update password set pass="' + password + '"where name="' + account + '"'
        conn.execute(edit)
        conn.commit()
        input("密碼更改完畢，請按ENTER返回主選單")
        print()


def deleteAccount():
    pressEnterLeave = False
    while True:
        account = input("請輸入要刪除的帳號(Enter==>停止輸入)")
        if account == "":
            pressEnterLeave = True
            break
        elif findSignalAccount(account, "checkAccountExist"):
            break
        else:
            print(account + " 帳號不存在!")
    if not pressEnterLeave:
        doubleCheck = input("確定刪除" + account + "的資料(Y/N)?")
        if doubleCheck == "Y" or doubleCheck == "y":
            delete = 'delete from password where name="' + account + '"'
            conn.execute(delete)
            conn.commit()
            input("帳號刪除完畢，請按ENTER返回主選單")
            print()
        else:
            deleteAccount()


def databaseInit():
    # check database if is exist
    if not os.path.isfile("Sqlite01.sqlite"):
        urllib.request.urlretrieve(
            "http://www1.pu.edu.tw/~yrjean/Sqlite01.sqlite", "Sqlite01.sqlite"
        )


databaseInit()
conn = sqlite3.connect("Sqlite01.sqlite")  # 建立資料庫線,若不存在,會先新增資料庫
while True:
    menu()
    choice = input("請輸入您的選擇：")
    print()
    if choice == "0":
        print("\n程式執行完畢！")
        break
    else:
        choices(choice)
```

```python
import json
import requests
import os
import hashlib
import sqlite3
from bs4 import BeautifulSoup
import json
with open('settings.json', 'r', encoding='utf8') as jfile:
    jdata = json.load(jfile)
conn = sqlite3.connect("DataBasePM25.sqlite")  # 建立資料庫連線,若不存在,則新建資料庫
cursor = conn.cursor()  # 建立 cursor物件

# 建立一個資料表(如果不存在)
sqlstr = """
CREATE TABLE IF NOT EXISTS TablePM25 ("no" INTEGER PRIMARY KEY AUTOINCREMENT
NOT NULL UNIQUE, "SiteName" TEXT NOT NULL, "PM25" INTEGER)
"""
cursor.execute(sqlstr)

# 細浮微粒資料(PM2.5) : 資料集代碼 aqx_p_02
url = f"https://data.epa.gov.tw/api/v2/aqx_p_02?api_key={jdata["api-key1"]}&format=json"

# 讀取網頁原始碼
html = requests.get(url).text  # .encode('utf-8')

if html == "請勿索取資料, 一分鐘呼API的次數不可大於1次":
    print(html)
    os._exit(0)

html = html.encode("utf-8")
# 判斷網頁是否更新
new_md5 = hashlib.md5(html).hexdigest()
old_md5 = ""
if os.path.exists("old_md5.txt"):
    with open("old_md5.txt", "r") as f:
        old_md5 = f.read()
with open("old_md5.txt", "w") as f:
    f.write(new_md5)

################
# 立即測試用,強設資料已更新
# old_md5 =
###########################

if new_md5 != old_md5:
    print("資料已更新,分析新資料...")
    print("=========================")

    sp = BeautifulSoup(html, "html.parser")
    # 將網頁內客轉換成JSON资料:dict
    jsondata = json.loads(sp.text)
    print(jsondata)
    print("------------------------")
    # 擷取JSON資料中監測站的PM2.5資料:list of dict
    sitesList = jsondata["records"]
    print(sitesList)

    print("=========================")
    print("清空資料庫, 寫入新資料")
    print("=========================")
    # 刪除資料表內客
    conn.execute("delete from TablePM25")
    conn.commit()

    n = 1
    for site in sitesList:
        SiteName = site["site"]
        PM25 = 0 if site["pm25"] == "" else int(site["pm25"])
        print(SiteName, PM25)
        # 新增一筆記錄
        sqlstr = "insert into TablePM25 values({},'{}',{})".format(n, SiteName, PM25)
        cursor.execute(sqlstr)
        n += 1
    conn.commit()  # 主動更新
else:
    print("資料未更新,從資料庫讀取...")
    print("=========================")
    cursor = conn.execute("select * from TablePM25")
    rows = cursor.fetchall()
    for row in rows:
        print(row[1], row[2])

conn.close()  # 關閉資料庫連
```

```python
from openpyxl import load_workbook
wb = load_workbook("Grades1101.xlsx")

sheet = wb[wb.sheetnames[0]]

exams = 0
for i in sheet["A"]:
    exams = exams + 1
    if i.value  =='系級':
        break
print("*** Exams at row:", exams)
for i in sheet[exams]:
    print(i.value)
print("================================================")

pct = exams -1
print("*** Pct at row:", pct)
for i in sheet[pct]:
    if type(i.value)==float:
        print("{:.0%}".format(i.value))
    else:
        print(i.value)
print("================================================")
```

```python
from openpyxl import load_workbook
wb = load_workbook("Grades1101.xlsx")

for sheets in wb.sheetnames:
    sheet = wb[sheets]
    
    print("================================================")
    print("*** sheet title:", sheet.title)

    # row by row, column by column
    rows = sheet.rows
    for row in rows:
        for col in row:
            if col.value != None:
                print(col.value,end=" ")
            else :
                print(" ",end="")
        print()
    
    print()
```

### [PT8-2]Excel_Analysis


```python
import urllib.request
from openpyxl import load_workbook

fileName = input("Excel filename:")
urllib.request.urlretrieve("http://www1.pu.edu.tw/~yrjean/" + fileName , fileName)
wb = load_workbook(fileName)
for sheets in wb.sheetnames:
    sheet = wb[sheets]
    print("================================================")
    print("*** Class:  ", sheet.title)
    Exams_columns = [];current_row = 0;Total_Pct = 0
    rows = sheet.rows
    for row in rows:
        exams = 0
        for col in row:
            exams+=1
            if type(col.value) == float:
                Exams_columns.append(exams)
                Total_Pct+=col.value
    print("*** Exams columns: ", Exams_columns)
    print("---------------------------------")
    for examname in Exams_columns:
        rows = str(chr(examname+64))
        for i in sheet[rows]:
            if i.value != None: 
                if type(i.value) == str and i.value != "缺":
                    print(i.value)
    print("*** Current Total Pct:  {:.0%}".format(Total_Pct))
```
